# <center>Attention Mechanism</center>

Today we are going to talk about the *Attention Mechanism* - that's is a core tecnique for LLM's.

Let's say that you want to translate the following sentence from English to Franch: "*The cat ate the mouse*" -> "*Le chat a mang√© la souris*"

If you want to translate this sentence one of the architectures that you can use is the **Encoder-Decoder Model**, but, sometimes the original sentence might not perfectly align with the translated sentence at each time step.

### An Example:

Take the sentence: "*Black cat ate the mouse*".
- In this example, the first english word is 'black' -> The first franch word is 'chat' which means 'cat' in english.




### *How can you train the model to focus more on the word 'cat' instead of the word 'black' at the first time step?*

To improve the translation you can add what is called the "*Attention Mechanism*" to the *encoder-decoder*.

- **Attention Mechanism** is a technique that allows a neural network to focus on specific parts of an input sequence.
    - This is done by assigning wights to differents parts of the input sequence with the most importante parts receiving the highest weights.
    - The weighted sum of the input sequence is than used as the input of the next layer of the neural network 

In a **Traditional RNN encoder-decoder** the model takes one word at a time as input and updates the hidden state and passes it on to the next time step.
> Only the final hidden state is passed on to the decoder for processing.



## Attention Model differs from a tradicional model

An attention model differs from a traditional sequence-to-sequence model in two main ways:
- **First:** *The encoder passes a lot more data to the decoder*
    - Instead of just passing the final hidden state number 3 to the decoder, the *encoder* passes all the hidden states from each timestep. This gives the decoder more context beyond the final hidden state.
    - The decoder uses all the hidden state information to translate the sentence.
- **Second:** *An Attention decoder takes an extra step before produciong output*
    - To focus on the most relevant parts of the input the 'Decoder' does the following: 
        - 1¬∫) It looks at the set of encoder hidden states that it received. Each encoder hidden state is associated with a certain word on the input sentence.
        - 2¬∫) It gives each hidden state a score.
        - 3¬∫) It multiplies each hidden state by its softmax score thus amplifying hidden states with high socres and downsizing hidden states with low scores.

Let's now connect all the pieces together, but, before moving on, let's define some notations:
- $\alpha$ represents the attention weights at each timestep.
- *H* represents the hidden state of encoder RNN at each timestep
- $H_d$ represents the hidden state of decoder RNN at each timestep.

The inversion of the "*black cat*" translation is clearly visible in the attention diagram, and 'ate' translates as two words 'a mang√©' in franch.

- During the attention step, we use the encoder hidden states and in the $H_4$ vector to calculate the context vector $a_4$ for this time step -> This is a weighted sum.
- We, then, concatenate $H_4$ and $a_4$ into one vector.
- This concatenated Vector is passed through a feedforward neural network, one trained jointly with the model.
- The output of the feedforward neural networks indicates the output word of this time step
- This process continues till the end of sentence token is generated by the decoder.

This is how you can use attention to improve the performance of the encoder-decoder architecture.
